```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de muestreo
fs = 20000                  # Frecuencia de muestreo (Hz)
t = np.linspace(0, 0.05, fs)  # Tiempo (50 ms)

# Señal 1 (X(t))
Az = 0.5                    # Amplitud
fx = 30                     # Frecuencia (Hz) dentro de 10–50 Hz
X_t = Az * np.cos(2 * np.pi * fx * t)

# Señal 2 (m(t))
An = 1.0                    # Amplitud
fm = 800                    # Frecuencia (Hz) dentro de 500–1000 Hz
m_t = An * np.cos(2 * np.pi * fm * t)

# Señal modulada
s_t = X_t * m_t

# Gráficas
plt.figure(figsize=(12, 9))

plt.subplot(3,1,1)
plt.plot(t, X_t, label="X(t)")
plt.title("Señal Portadora X(t) = 0.5 cos(2π·30t)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid()

plt.subplot(3,1,2)
plt.plot(t, m_t, label="m(t)", color="orange")
plt.title("Señal Moduladora m(t) = 1.0 cos(2π·800t)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid()

plt.subplot(3,1,3)
plt.plot(t, s_t, label="s(t)", color="green")
plt.title("Señal Modulada s(t) = X(t) * m(t)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
